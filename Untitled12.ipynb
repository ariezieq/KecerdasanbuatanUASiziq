```python
# ================================================
#   PENGOLAHAN CITRA  OPERASI GEOMETRI LENGKAP
# ================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ------------------------------------------------
# 1. Upload Gambar
# ------------------------------------------------
print("Silakan upload gambar...")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
img = cv2.imread(file_name)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))
plt.title("Citra Asli")
plt.imshow(img)
plt.axis("off")
```

```python
# ------------------------------------------------
# 2. Translasi (geser 50 px ke kanan, 30 px ke bawah)
# ------------------------------------------------

rows, cols = img.shape[:2]
M_translate = np.float32([[1, 0, 50],  # geser kanan 50
                           [0, 1, 30]]) # geser bawah 30

translated = cv2.warpAffine(img, M_translate, (cols+50, rows+30))

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.title("Sebelum Translasi")
plt.imshow(img); plt.axis("off")
plt.subplot(1,2,2); plt.title("Setelah Translasi")
plt.imshow(translated); plt.axis("off")
```

```python
# ------------------------------------------------
# 3. Scaling (1.5x dan 0.5x)
# ------------------------------------------------

scale_up = cv2.resize(img, None, fx=1.5, fy=1.5)
scale_down = cv2.resize(img, None, fx=0.5, fy=0.5)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Asli"); plt.imshow(img); plt.axis("off")
plt.subplot(1,3,2); plt.title("Diperbesar 1.5x"); plt.imshow(scale_up); plt.axis("off")
plt.subplot(1,3,3); plt.title("Dikecilkan 0.5x"); plt.imshow(scale_down); plt.axis("off")
```

```python
# ------------------------------------------------
# 4. Rotasi
# ------------------------------------------------

M_rot45 = cv2.getRotationMatrix2D((cols/2, rows/2), -45, 1)  # searah jarum jam
rot45 = cv2.warpAffine(img, M_rot45, (cols, rows))

M_rot90 = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)   # berlawanan
rot90 = cv2.warpAffine(img, M_rot90, (cols, rows))

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Asli"); plt.imshow(img); plt.axis("off")
plt.subplot(1,3,2); plt.title("Rotasi 45 CW"); plt.imshow(rot45); plt.axis("off")
plt.subplot(1,3,3); plt.title("Rotasi 90 CCW"); plt.imshow(rot90); plt.axis("off")
```

```python
# ------------------------------------------------
# 5. Refleksi
# ------------------------------------------------

flip_x = cv2.flip(img, 0)   # sumbu X (atas-bawah)
flip_y = cv2.flip(img, 1)   # sumbu Y (kiri-kanan)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Asli"); plt.imshow(img); plt.axis("off")
plt.subplot(1,3,2); plt.title("Refleksi X-axis"); plt.imshow(flip_x); plt.axis("off")
plt.subplot(1,3,3); plt.title("Refleksi Y-axis"); plt.imshow(flip_y); plt.axis("off")
```

```python
# ------------------------------------------------
# 6. Warping (Perspective Transform)
# ------------------------------------------------

pts1 = np.float32([[0,0], [cols,0], [0,rows], [cols,rows]])  # persegi asli
pts2 = np.float32([[50,0], [cols-50,0], [0,rows], [cols,rows]])  # trapesium

M_persp = cv2.getPerspectiveTransform(pts1, pts2)
warp = cv2.warpPerspective(img, M_persp, (cols, rows))

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.title("Asli"); plt.imshow(img); plt.axis("off")
plt.subplot(1,2,2); plt.title("Warping (Trapesium)"); plt.imshow(warp); plt.axis("off")
```

```python
# ------------------------------------------------
# 7. Affine Transformation
# ------------------------------------------------

pts1_aff = np.float32([[0,0], [cols-1,0], [0,rows-1]])
pts2_aff = np.float32([[30,30], [cols+20,50], [50,rows]])

M_affine = cv2.getAffineTransform(pts1_aff, pts2_aff)
affine = cv2.warpAffine(img, M_affine, (cols+100, rows+100))

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.title("Asli"); plt.imshow(img); plt.axis("off")
plt.subplot(1,2,2); plt.title("Affine Transform"); plt.imshow(affine); plt.axis("off")

```

```python
# Affine: kombinasi translasi + rotasi + scaling (Colab-ready)
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
print("Upload gambar (jpg/png):")
uploaded = files.upload()
fname = next(iter(uploaded))
img = cv2.imread(fname)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rows, cols = img.shape[:2]

# --- Parameter transformasi (ubah sesuai kebutuhan) ---
angle_deg = 30        # rotasi (derajat). Positif = CCW (counter-clockwise)
scale_x = 1.2         # skala pada sumbu x
scale_y = 1.2         # skala pada sumbu y (jika ingin non-uniform, tetapkan berbeda)
tx = 80               # translasi ke kanan (pixel)
ty = 40               # translasi ke bawah (pixel)

# --- Hitung matriks 2x2 (rotasi * scaling) ---
theta = np.deg2rad(angle_deg)
cos_t = np.cos(theta)
sin_t = np.sin(theta)

# Jika ingin: M2 = R * S  (scale lalu rotate)
M2 = np.array([[cos_t * scale_x, -sin_t * scale_y],
               [sin_t * scale_x,  cos_t * scale_y]], dtype=np.float32)

# Gabungkan dengan translasi menjadi matriks affine 2x3
M_aff = np.hstack([M2, np.array([[tx],[ty]], dtype=np.float32)])  # shape (2,3)

# --- Agar hasil tidak terpotong: hitung bounding box transformasi dari 4 sudut ---
corners = np.array([[0,0,1],
                    [cols,0,1],
                    [0,rows,1],
                    [cols,rows,1]], dtype=np.float32).T  # shape (3,4)

# Konversi M_aff menjadi 3x3 untuk perkalian homogen
M3 = np.vstack([M_aff, [0,0,1]])
trans_corners = (M3 @ corners).T[:, :2]  # shape (4,2)

min_x, min_y = trans_corners.min(axis=0)
max_x, max_y = trans_corners.max(axis=0)

# Jika ada nilai negatif, kita perlu menggeser seluruh transform agar koordinat >= 0
shift_x = -min_x if min_x < 0 else 0
shift_y = -min_y if min_y < 0 else 0

# Perbarui translasi pada matriks affine dengan shift supaya semua positif
M_aff_shifted = M_aff.copy()
M_aff_shifted[0,2] += shift_x
M_aff_shifted[1,2] += shift_y

# Tentukan ukuran kanvas output yang cukup besar
out_w = int(np.ceil(max_x + shift_x))
out_h = int(np.ceil(max_y + shift_y))

# Jaga minimal ukuran agar tidak terlalu kecil
out_w = max(out_w, cols + int(abs(tx)) + 10)
out_h = max(out_h, rows + int(abs(ty)) + 10)

# Terapkan transformasi affine
result = cv2.warpAffine(img, M_aff_shifted, (out_w, out_h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT, borderValue=(0,0,0))

# --- Tampilkan ---
plt.figure(figsize=(12,6))
plt.subplot(1,2,1); plt.title("Citra Asli"); plt.imshow(img); plt.axis('off')
plt.subplot(1,2,2); plt.title(f"Affine: rot {angle_deg}, sx={scale_x}, sy={scale_y}, tx={tx}, ty={ty}"); plt.imshow(result); plt.axis('off')
plt.show()

# --- Tampilkan matriks affine yang dipakai ---
print("Matriks affine (2x3) yang dipakai:")
print(M_aff_shifted)
```

**Warping (perspective transform)** merupakan transformasi yang dapat mengubah citra secara lebih fleksibel karena mampu menghasilkan efek perspektif, yaitu membuat garis-garis yang semula sejajar menjadi tidak sejajar. Transformasi ini menggunakan matriks 33 sehingga bentuk citra dapat berubah seperti tampak menyempit, melebar, atau menjadi bentuk trapesium.

**affine transformation **menggunakan matriks 23 dan mempertahankan sifat
keselarasan garis (parallelism preserved), sehingga meskipun dapat melakukan
translasi, rotasi, scaling, dan shear, transformasi ini tidak dapat menghasilkan efek perspektif. Oleh karena itu, warping memberikan distorsi yang lebih realistis seperti pengambilan gambar dari sudut kamera, sedangkan affine hanya melakukan perubahan geometri dasar tanpa mengubah perspektif citra.
